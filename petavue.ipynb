```python
### Combined the 2016 and 2017 data into 1 dataframe
df_2016_17 = pd.concat([df_2016_airport, df_2017_airport])
```

```python
### Make a list of desired columns from the flight dataset
desired_columns = ['FlightDate', 'Quarter', 'Year', 'Month', 'DayofMonth', 'DepTime', 'DepDel15', 'CRSDepTime', 'DepDelayMinutes', 'Origin', 'Dest', 'ArrTime', 'CRSArrTime', 'ArrDel15', 'ArrDelayMinutes']
```

```python
df_2016_17_filtered.columns
```

```python
df_2016_17_filtered['FlightDate'] = pd.to_datetime(df_2016_17_filtered['FlightDate'])

# Convert ArrTime column to datetime format
df_2016_17_filtered['ArrTime_new'] = pd.to_datetime(df_2016_17_filtered['ArrTime'], format='%H%M', errors='coerce')

# Create a new column arrival_time by rounding off the ArrTime to the nearest hour
df_2016_17_filtered['arrival_time'] = df_2016_17_filtered['ArrTime_new'].dt.round('H').dt.strftime('%H%M')

# Create a new column arrival_date using the FlightDate
df_2016_17_filtered['arrival_date'] = df_2016_17_filtered['FlightDate'].dt.strftime('%Y-%m-%d')
```

```python
def create_dataframe_from_file(json_filepath):
    json_data = json.loads(open(json_filepath).read())
    return create_dataframe_from_json(json_data)

def load_and_concatenate_data(directory_path, airport_codes):
    dfs = []
    
    for airport_code in airport_codes:
        for year in range(2016, 2018):
            for month in range(1, 13):
                json_filename = f'{airport_code}-001/{airport_code}/{year}-{month}.json'
                json_filepath = os.path.join(directory_path, json_filename)

                if os.path.exists(json_filepath):
#                     print(f"Loading {json_filepath}")
                    df = create_dataframe_from_file(json_filepath)
                    df['airport'] = airport_code
                    dfs.append(df)
                else:
                    print(f"File not found: {json_filepath}")

    if not dfs:
        print("No DataFrames to concatenate.")

    concatenated_df = pd.concat(dfs, ignore_index=True)
    return concatenated_df

def create_dataframe_from_json(json_object):
    hourly_data_list = []

    for daily_entry in json_object['data']['weather']:
        date = daily_entry['date']
        
        for hourly_entry in daily_entry['hourly']:
            hourly_data = {
                'WindSpeedKmph': hourly_entry['windspeedKmph'],
                'WindDirDegree': hourly_entry['winddirDegree'],
                'WeatherCode': hourly_entry['weatherCode'],
                'precipMM': hourly_entry['precipMM'],
                'Visibility': hourly_entry['visibility'],
                'Pressure': hourly_entry['pressure'],
                'Cloudcover': hourly_entry['cloudcover'],
                'DewPointF': hourly_entry['DewPointF'],
                'WindGustKmph': hourly_entry['WindGustKmph'],
                'tempF': hourly_entry['tempF'],
                'WindChillF': hourly_entry['WindChillF'],
                'Humidity': hourly_entry['humidity'],
                'date': date,
                'time': hourly_entry['time'],
            }
            
            hourly_data_list.append(hourly_data)

    df = pd.DataFrame(hourly_data_list)
    return df


directory_path = '/kaggle/input/petavue-data1'
airport_codes = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']

final_df = load_and_concatenate_data(directory_path, airport_codes)
```
